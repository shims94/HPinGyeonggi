HPinGyeunggi는 경기도 소재 의료시설에 대한 정보를 제공하며, 경기도 소재 의료시설의 이름과 위치정보, 응급센터등의 특별시설 여부에 대해 정보를 제공하는 프로그램입니다.

   경기데이터 드림 제공

```python
pip install folium

import pandas as pd
import folium
import numpy as np
from folium.plugins import MarkerCluster

# 데이터셋 로드
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vSIs5XJ0xXwdGKUQar6LCakW_KGOA4seQbTzfc0pr8gBhFJLVVtTvp1Y-LN2R0Zm5NPpBWg3DTeUTpt/pub?gid=540812427&single=true&output=csv')

df.head(3)
```

```python
df = df.rename(columns={'병원명/센터명': '병원또는센터명'})

# 응급의료지원센터여부	전문응급의료센터여부
# 전문응급센터전문분야	권역외상센터여부	지역외상센터여부의 NaN을 fillna()함수로 0으로 채워줌

df['응급의료지원센터여부'] = df['응급의료지원센터여부'].fillna('0')
df['전문응급의료센터여부'] = df['전문응급의료센터여부'].fillna('0')
df['전문응급센터전문분야'] = df['전문응급센터전문분야'].fillna('0')
df['권역외상센터여부'] = df['권역외상센터여부'].fillna('0')
df['지역외상센터여부'] = df['지역외상센터여부'].fillna('0')
df
```

```python
# 메인페이지 시각화부분
# 초기 맵
map_gyeonggi = folium.Map(location=[37.291887, 126.996340	],
               zoom_start=10,
               width=500,
               height=500)

# 결측치를 삭제한 사본데이터프레임
df_nan_delete = df
df_nan_delete = df_nan_delete.dropna(subset=['소재지지번주소'], how='any', axis=0)

# 업무 구분명이 지역센터나 기관일경우 녹색 하트마커, 그이외(광역) 빨강 스타마커
for index, row in df_nan_delete.iterrows():
    if row['업무구분명'] == '지역센터' or row['업무구분명'] == '지역기관':
        folium.Marker(location = [row['위도'], row['경도']],
                      popup=row['병원또는센터명'],
                      icon=folium.Icon(color='green',icon='heart')
                     ).add_to(map_gyeonggi)
    else:
        folium.Marker(location = [row['위도'], row['경도']],
                      popup=row['병원또는센터명'],
                    icon=folium.Icon(color='yellow',icon='star')
                     ).add_to(map_gyeonggi)
map_gyeonggi
```
